# - 크롤링

```python
import requests
from bs4 import BeautifulSoup
```

```python
source = requests.get("https://www.royalcanin.com/kr/dogs/products/retail-products/small-coat-care-dry-dog-food").content

soup = BeautifulSoup(source, "html.parser")
key = soup.findAll("div", {'class':'rc-full-width'})
```

```python
for name in key:
    print(name.get_text())
```

# - 엑셀로 저장

```python
import openpyxl
```

```python
a = openpyxl.load_workbook('a.xlsx')

load = a['a']
# sheet = a.active
# sheet.cell()
# print(load.cell(:).value)
get_cells = load['A1':'B1']
for row in get_cells:
    for cell in row:
        print(cell.value)
```

```python
sheet = a.worksheets[0]
```

```python
# 찾으려는 문자가 있다면 True를 return 한다.
def findExistString(searchString, text):
    if searchString in text: return True
    else: return False
```

```python
# 시트의 각 행을 순서대로 추출하면서 실행한다.
for row in sheet.rows:
    if findExistString("오리", row[1].value):
        print(row[0].value)
    else :
        print('없어요')
```

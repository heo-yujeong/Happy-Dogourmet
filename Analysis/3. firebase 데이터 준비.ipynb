```python
import numpy as np
import pandas as pd
import os
```

# 1. 사료데이터

```python
saryo = os.path.join('data/사료데이터합본_태깅추가.xlsx')
```

```python
df_saryo = pd.read_excel(saryo, sheet_name='Sheet1', dtype={'no':str,'name':str, 'link':str, 'brand':str, 'material':str, 'tagging':str, 'clust':str})
```

```python
for i in range(14):
    df_saryo_slice = df_saryo[df_saryo['clust']==str(i)]
    saryo_data_slice = df_saryo_slice[['no','name','link','brand','material','tagging']]
    file_name = 'data/saryo_' + str(i) +'.csv'
    saryo_data_slice.to_csv(file_name, mode='w', encoding='utf-8-sig')
```
